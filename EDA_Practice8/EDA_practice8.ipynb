## 20. Classify Movies based on rating ['Excellent', 'Good', 'Moderate' and 'Bad']

```python
def func(x):
    if x > 8:
        return 'Excellent'
    elif 7 <= x < 8:
        return 'Good'
    elif 5 <= x < 7:
        return 'Moderate'
    else:
        return 'Bad'
    
```

```python
data['Class'] = data['Rating'].apply(func)
```

## 22. Number of unique Genres

```python
genres = []
for value in data['Genre']:
    for val in value.split(','):
        if val not in genres:
            genres.append(val)

print("Unique Genres:", len(genres))
print("Genres:", genres)
```

## 23. How many films of each Genre are there in the dataset?

```python
genre_cnt = {}
for genre in genres:
    genre_cnt[genre] = sum(data['Genre'].str.contains(genre))

plt.title("Movie count by Genre")
plt.xlabel("Movie Count", fontsize=13)
plt.ylabel("Genre", fontsize=13)
sns.barplot(x=genre_cnt.values(), y=genre_cnt.keys())
plt.show()
```
